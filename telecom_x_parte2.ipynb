🎯 Metodo recomendado: Qui-cuadrado (Chi-Square Test)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2

# Crea copia
df_rank = df.copy()

# Convertir Churn para binario
df_rank["Churn"] = df_rank["Churn"].map({"Yes": 1, "No": 0})

# Separar variables categóricas
categorical_cols = df_rank.select_dtypes(include=["object"]).columns

# Aplicar Label Encoding temporario
le = LabelEncoder()
for col in categorical_cols:
    df_rank[col] = le.fit_transform(df_rank[col])

# Separar X e y
X_cat = df_rank[categorical_cols]
y = df_rank["Churn"]

# Aplicar teste Qui-cuadrado
chi_scores, p_values = chi2(X_cat, y)

# Criar DataFrame com ranking
chi_df = pd.DataFrame({
    "Variable": categorical_cols,
    "Chi2 Score": chi_scores,
    "p-value": p_values
})

# Ordenar por importancia
chi_df = chi_df.sort_values(by="Chi2 Score", ascending=False)

chi_df
```

```python
significant_vars = chi_df[chi_df["p-value"] < 0.05]

print("\n✅ Variables significativas (p < 0.05):")
display(significant_vars)
```

```python
df = df.drop(columns=["Gender", "PhoneService"])
```

## 🎯 Split

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

2.Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

3.One-Hot Encoding

```python
# Aplicar One-Hot Encoding en treno
X_train = pd.get_dummies(X_train, drop_first=True)

# Aplicar en teste
X_test = pd.get_dummies(X_test, drop_first=True)

# Garantir que treno y teste tengan las mismas columnas
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

##Entrenamiento

Modelo 1 — Regresion logistica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred_log_normal = log_model.predict(X_test)

print("Regresion logistica")
print(classification_report(y_test, y_pred_log_normal))
```

Modelo 2 — Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, class_weight="balanced")
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)

print(classification_report(y_test, y_pred_log))
```

Análisis de la Importancia de Variables: Predicción de Cancelación (Churn)
En el modelo de Regresión Logística, la importancia de cada variable se evalúa a través de sus coeficientes. Estos coeficientes no solo nos indican qué características tienen el mayor peso en la decisión del cliente, sino también la dirección de ese impacto: si aumentan la probabilidad de que el cliente abandone la empresa (riesgo) o si, por el contrario, ayudan a fidelizarlo (retención).

Al analizar los coeficientes extraídos del modelo, podemos dividir los hallazgos en dos grandes grupos:

1. Factores de Riesgo (Variables que impulsan la cancelación)
Estas son las características asociadas a coeficientes positivos altos. Cuando un cliente posee estos atributos, la probabilidad de que cancele el servicio (Churn = Yes) aumenta considerablemente:

Contratos mensuales (Contract_Month-to-month): Es consistentemente el mayor predictor de abandono. Los clientes sin compromisos a largo plazo tienen total libertad para cambiar de proveedor ante la menor insatisfacción o mejor oferta de la competencia.

Servicio de Internet por Fibra Óptica (InternetService_Fiber optic): Aunque suele ser un servicio superior, el modelo indica que estos clientes son más propensos a irse. Esto sugiere que podría haber problemas de estabilidad en el servicio, expectativas no cumplidas o que el costo de la fibra óptica no está justificado por la calidad percibida.

Método de pago electrónico (PaymentMethod_Electronic check): Los clientes que utilizan cheques electrónicos muestran una mayor tasa de cancelación en comparación con los que tienen pagos automáticos (como tarjetas de crédito o débito automático).

Cargos mensuales (ChargesMonthly): A medida que aumenta el costo mensual del servicio, también lo hace la fricción con el cliente, incrementando el riesgo de que busque alternativas más económicas.

2. Factores de Retención (Variables que previenen la cancelación)
Estas variables presentan coeficientes negativos fuertes, lo que significa que actúan como "anclas" que mantienen al cliente en la empresa (Churn = No):

Antigüedad del cliente (Tenure): Es el factor de retención más poderoso. Cuanto más tiempo lleva un cliente en la empresa, menor es su probabilidad de cancelación. Esto demuestra que la lealtad se consolida con el tiempo.

Contratos a largo plazo (Contract_Two year / One year): Los clientes con contratos de uno o dos años están altamente fidelizados. El compromiso contractual reduce drásticamente el abandono.

Servicios de soporte y seguridad (TechSupport & OnlineSecurity): Los clientes que contratan servicios adicionales, como soporte técnico premium o seguridad en línea, rara vez cancelan. Esto indica que perciben un alto valor en la atención al cliente y se sienten más respaldados por la empresa.

Tener dependientes o pareja (Dependents / Partner): Los hogares familiares o con múltiples personas tienden a buscar estabilidad y son menos propensos a cambiar de proveedor de telecomunicaciones con frecuencia.

💡 Conclusión Estratégica para el Negocio
El modelo de Regresión Logística demuestra que la cancelación no es un evento aleatorio, sino que está fuertemente ligado a la falta de compromisos a largo plazo y la insatisfacción potencial con servicios costosos (como la fibra óptica).

Recomendaciones:
Para reducir el Churn, la empresa debería enfocar sus estrategias en incentivar la migración de clientes con contratos mensuales hacia planes anuales (quizás ofreciendo descuentos) y revisar urgentemente la calidad y el precio del servicio de Fibra Óptica. Además, regalar u ofrecer a bajo costo servicios como Soporte Técnico u Online Security durante los primeros meses podría actuar como una excelente barrera de retención.
